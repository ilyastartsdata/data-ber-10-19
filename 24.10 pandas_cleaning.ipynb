```python
def detect_low_variance(df, column, perc):
    min_value = df[column].min()
    perc_value = np.percentile(df[column], perc)
    return min_value == perc_value
```

```python
detect_low_variance(df, "barrels08", 90)
```

```python
df.select_dtypes(include = [np.number]).columns # display the columns
```

```python
low_variance = [col for col in df.select_dtypes(include = [np.number]).columns if detect_low_variance(df,col,90)]
```

```python
df.drop(low_variance, axis = 1)
```

## 3. Outliers

Using IQR

```python
df.describe().transpose()
```

```python
df.describe().transpose()["25%"]
```

```python
df_desc = df.describe().transpose()
df_desc = df_desc.assign(IQR = df_desc["75%"] - df_desc["25%"])
```

```python
df_desc
```

```python
df_desc.index
```

```python
outliers = pd.DataFrame(columns = df.columns)
for col in df_desc.index:
    iqr = df_desc.loc[col, "IQR"]
    cutoff = iqr * 6
    lower_threshold = df_desc.loc[col, "25%"] - cutoff
    upper_threshold = df_desc.loc[col, "75%"] + cutoff
    results = df[(df[col] < lower_threshold) | (df[col] > upper_threshold)].copy()
    results = results.assign(outlier = col)
    outliers = pd.concat([outliers, results], ignore_index = True, sort = False)
```

```python
outliers
```

```python
# count how many outliers
outliers.groupby("outlier")["outlier"].count()
```

```python
def check_drop_duplicates(#. a dataframe, # a list of columns in that dataframe):
    #1. create a copy of the data frame
    #2. drop the duplicate rows from that copy given the list
    #3. tell user how many rows would get dropped and ask if i should proceed
        #4.1 if yes, drop the rows on df (inplace)
        #4.2 if no, do not drop, but then ask user if i should return the rows that would get dropped
            #5.1 if no, exit
            #5.2 if yes, retrieve the rows that would get dropped, and return them as a DataFrame
```

```python
def check_drop_duplicates(df, cols):
    
    #1. create a copy of the data frame
    
    copy = df.copy()
    
    #2. drop the duplicate rows from that copy given the list
    
    copy = copy.drop_duplicates(cols) # this could be done in one step
    
    #3. tell user how many rows would get dropped and ask if i should proceed
    
    before = len(df)
    after = len(copy)
    drop_choice = None
    while drop_choice not in ["y","n"]:
        drop_choice = input(f'{before-after} duplicated rows identified. Do u want me to drop them? (y/n)')
    
        #4.1 if yes, drop the rows on df (inplace)
        
    if drop_choice == "y":    
        df.drop_duplicates(cols, inplace = True)
        return
    
        #4.2 if no, do not drop, but then ask user if i should return the rows that would get dropped
    print('Aborting')
    return_choice = None
    while return_choice not in ["y","n"]:
        return_choice = input('Do u want to get the rows, that is going to be dropped? (y/n)')
    
    if return_choice == "y":
        return df.loc[list(set(df.index) - set(copy.index))]
            #5.1 if no, exit
            #5.2 if yes, retrieve the rows that would get dropped, and return them as a DataFrame
```
