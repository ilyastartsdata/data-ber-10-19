# Bitcoin market value for last 30 days, vs. USD, deliver a DF and a Plot

```python
base = 'https://api.coingecko.com/api/v3/'
```

```python
url = base + 'coins/bitcoin/market_chart?vs_currency=usd&days=30'
```

```python
result = requests.get(url)
bitcoin_data = result.json()
```

```python
bitcoin_data
```

```python
df_bitcoin = pd.DataFrame(bitcoin_data["prices"], columns = ["time","price"])
df_bitcoin.head()
```

```python
df_bitcoin["date"]=pd.to_datetime(df_bitcoin["time"],unit='ms')
```

```python
df_bitcoin = df_bitcoin.drop(["time"],axis=1)
```

```python
df_bitcoin
```

```python
df_bitcoin.set_index("date",inplace=True)
```

```python
df_bitcoin
```

```python
df_bitcoin.plot()
```

# Top Ten Coins Report

```python
url3 = base + 'coins/list'
```

```python
results3 = requests.get(url3)
top_ten_data = results3.json()
```

```python
top_ten_data
```

```python
top_df = pd.DataFrame(top_ten_data)
top_df.head()
```

```python
data = []

for i in top_df["id"]:
    base = 'https://api.coingecko.com/api/v3/coins/'
    url4 = base + i + '/market_chart?vs_currency=usd&days=300'
    result4 = requests.get(url4)
    top_ten_fun = result4.json()
    data_top_ten = pd.DataFrame(top_ten_fun['prices'], columns = ['date_time','price'])
    data_top_ten["file_name"] = i
    data.append(data_top_ten)
    print(i)
```
