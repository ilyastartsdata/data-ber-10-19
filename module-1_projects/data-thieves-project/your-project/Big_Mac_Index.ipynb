### Importing the currency exchange rate for 2018-07-31

and merging the dataframes 

```python
# Importing the csv

csv_20180731 = pd.read_csv('currency_exchange_20180731.csv', sep = ';')
csv_20180731.columns = csv_20180731.columns.str.lower()
csv_20180731
```

```python
# Creating the new dataframe, where only the one date is presented

df_20180731 = df_international[df_international['date'] == '2018-07-31']
```

```python
df_20180731
```

```python
# Merging two dataframes based on 'country' column

united_20180731 = pd.merge(csv_20180731, df_20180731, on='country')
```

```python
# Step 2 in calculation of BMI:
# Dividing relation column by
# the units_per_usd column and
# subtracting the result from 1
# in order to get the percentage

united_20180731['1_minus_difference'] = (1 - united_20180731['relation']/united_20180731['units_per_usd'])
```

```python
# Evaluating the result

united_20180731['evaluation'] = np.where(united_20180731['1_minus_difference'] > 0, 'undervalued', 'overvalued')
```

```python
# Creating the new 'clean' dataframe

result_20180731 = united_20180731[['date','country','local_price','USA_price','relation','units_per_usd','1_minus_difference','evaluation']]
```

```python
# Adding the column in order to plot it later

result_20180731['price_in_usd'] = result_20180731['local_price']/result_20180731['units_per_usd']
```

```python
result_20180731
```

```python
# Plot

result_20180731.plot(x='country',y='price_in_usd',figsize=(20, 10),kind='bar',title = 'Price for Big Mac in USD across the world',rot=0)
```

```python
result_20180731.groupby('evaluation')['evaluation'].count().plot.pie(autopct='%.2f',figsize=(10,10),title='International currency state: 2018-07-31')
```

## Part 2: Evaluating the usage of EU-rating instead of rating for each country

While doing the first part by comparing prices for Big Macs in different countries vs the price in the USA, I bumped into some stuff, which I firstly didn't understand. After 2018-01-31 there was no more data separatily for countries, which use EUR as their national currency, only in form of general EU-rating...
